Observamos un numero negativo en symboling, observamos outliers

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['sulphates'])
plt.title('Boxplot of Symboling')
plt.ylabel('Symboling')
plt.show()
```

Se observan posibles outliers en el extremo superior de la distribución de sulphates.

```python
#Obtenemos Outliers

Q1 = df['sulphates'].quantile(0.25)
Q3 = df['sulphates'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
outliers = df[(df['sulphates'] < lower_bound) | (df['sulphates'] > upper_bound)]
```

```python
outliers
```

No se observan valores extremos o ngeativos, se atribuyen outliers a propiedades particulares de algunos vinos.

```python
# Mapa de calor de la matriz de correlación
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de la Matriz de Correlación')
plt.show()
```

La matriz de correlación muestra que la variable más relacionada con la calidad del vino es el alcohol, con una correlación positiva moderada. También destacan sulphates y citric_acid como variables ligeramente asociadas a mayor calidad. Por el contrario, volatile_acidity tiene una correlación negativa, indicando que niveles altos se asocian con menor calidad.

```python
# Histograma de la distribución de la variable 'quality'
plt.figure(figsize=(8, 6))
sns.countplot(x='quality', data=df, palette='viridis')
plt.title('Distribución de la Calidad del Vino')
plt.xlabel('Calidad (Score)')
plt.ylabel('Número de Vinos')
plt.show()
```

Observaciones clave:
- La distribución de la calidad del vino está sesgada hacia los valores medios.
- Los vinos con puntuaciones de 5, 6 y 7 son los más comunes, representando la mayoría de los datos.
- Hay menos vinos con puntuaciones muy bajas (3 y 4) o muy altas (8).
- La puntuación más frecuente es 5 o 6, lo que sugiere que la mayoría de los vinos en este dataset tienen una calidad media.

 Implicaciones:
- Dado que la mayoría de los vinos tienen una calidad media, el dataset está desequilibrado en términos de calidad.
- Esto es importante a considerar si se va a construir un modelo de clasificación para predecir la calidad, ya que las clases de menor y mayor calidad tendrán menos ejemplos para entrenar.

```python
# Gráfico de dispersión entre 'alcohol' y 'quality'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df, alpha=0.6)
plt.title('Alcohol vs Calidad del Vino')
plt.xlabel('Contenido de Alcohol')
plt.ylabel('Calidad del Vino')
plt.show()
```

El gráfico de dispersión entre 'alcohol' y 'quality' sugiere una tendencia positiva: a medida que aumenta el contenido de alcohol,
tiende a aumentar la calidad del vino. Sin embargo, la relación no es perfectamente lineal y existe dispersión,
lo que indica que el alcohol es un factor importante pero no el único determinante de la calidad.
Vinos con alto contenido de alcohol muestran una mayor concentración en las calidades superiores.

```python
# Gráfico de violín para 'fixed acidity' vs 'quality'
plt.figure(figsize=(8, 6))
sns.violinplot(x='quality', y='fixed_acidity', data=df, palette='pastel')
plt.title('Acidez Fija vs Calidad del Vino')
plt.xlabel('Calidad del Vino')
plt.ylabel('Acidez Fija')
plt.show()
```

El gráfico de violín entre la acidez fija y la calidad del vino muestra cómo la distribución de la acidez fija varía para cada nivel de calidad.
Las formas de los violines sugieren que:
-Para calidades bajas (ej. 3 y 4), la distribución de la acidez fija es más concentrada en valores más bajos.
-A medida que aumenta la calidad (ej. 5, 6, 7, 8), la distribución de la acidez fija tiende a desplazarse hacia valores ligeramente más altos o a tener una mayor dispersión.



#Diagnóstico y Selección de Problema:

**Características Principales:**
El dataset incluye variables fisicoquímicas del vino (alcohol, acidez, pH, etc.) y una variable objetivo de calidad (escala 3-8), con mayor concentración en valores intermedios (5-7). El alcohol muestra la correlación positiva más fuerte con la calidad, mientras la acidez volátil tiene correlación negativa. Se detectan outliers en 'sulphates', pero son valores válidos.

**Desafíos Clave:**
El desequilibrio en las clases de calidad (pocos ejemplos en puntuaciones extremas) complica la predicción precisa, especialmente para clasificación. La naturaleza discreta y subjetiva de la calidad plantea dudas sobre abordarla como clasificación o regresión. Además, las múltiples características (11) requieren selección para evitar sobreajuste, y las correlaciones no implican causalidad directa.

**Elección de Clasificación:**
Se prioriza como problema de clasificación (predecir categorías de calidad) por su aplicabilidad práctica y para identificar factores clave diferenciadores. El desequilibrio se manejará con técnicas como oversampling o métricas específicas (F1-score), aprovechando que el alcohol y acidez volátil son predictores claros, pero integrando otras variables para capturar la complejidad multifactorial de la calidad del vino.
