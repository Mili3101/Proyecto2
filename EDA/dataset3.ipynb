```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['sulphates'])
plt.title('Boxplot of Symboling')
plt.ylabel('Symboling')
plt.show()
```

Se observan posibles outliers en el extremo superior de la distribuciÃ³n de sulphates.

```python
#Obtenemos Outliers

Q1 = df['sulphates'].quantile(0.25)
Q3 = df['sulphates'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
outliers = df[(df['sulphates'] < lower_bound) | (df['sulphates'] > upper_bound)]
```

```python
outliers
```

No se observan valores extremos o ngeativos, se atribuyen outliers a propiedades particulares de algunos vinos.

```python
# Mapa de calor de la matriz de correlaciÃ³n
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Mapa de Calor de la Matriz de CorrelaciÃ³n')
plt.show()
```

La matriz de correlaciÃ³n muestra que la variable mÃ¡s relacionada con la calidad del vino es el alcohol, con una correlaciÃ³n positiva moderada. TambiÃ©n destacan sulphates y citric_acid como variables ligeramente asociadas a mayor calidad. Por el contrario, volatile_acidity tiene una correlaciÃ³n negativa, indicando que niveles altos se asocian con menor calidad.

```python
# Histograma de la distribuciÃ³n de la variable 'quality'
plt.figure(figsize=(8, 6))
sns.countplot(x='quality', data=df, palette='viridis')
plt.title('DistribuciÃ³n de la Calidad del Vino')
plt.xlabel('Calidad (Score)')
plt.ylabel('NÃºmero de Vinos')
plt.show()
```

Observaciones clave:
- La distribuciÃ³n de la calidad del vino estÃ¡ sesgada hacia los valores medios.
- Los vinos con puntuaciones de 5, 6 y 7 son los mÃ¡s comunes, representando la mayorÃ­a de los datos.
- Hay menos vinos con puntuaciones muy bajas (3 y 4) o muy altas (8).
- La puntuaciÃ³n mÃ¡s frecuente es 5 o 6, lo que sugiere que la mayorÃ­a de los vinos en este dataset tienen una calidad media.

 Implicaciones:
- Dado que la mayorÃ­a de los vinos tienen una calidad media, el dataset estÃ¡ desequilibrado en tÃ©rminos de calidad.
- Esto es importante a considerar si se va a construir un modelo de clasificaciÃ³n para predecir la calidad, ya que las clases de menor y mayor calidad tendrÃ¡n menos ejemplos para entrenar.

```python
# GrÃ¡fico de dispersiÃ³n entre 'alcohol' y 'quality'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df, alpha=0.6)
plt.title('Alcohol vs Calidad del Vino')
plt.xlabel('Contenido de Alcohol')
plt.ylabel('Calidad del Vino')
plt.show()
```

El grÃ¡fico de dispersiÃ³n entre 'alcohol' y 'quality' sugiere una tendencia positiva: a medida que aumenta el contenido de alcohol,
tiende a aumentar la calidad del vino. Sin embargo, la relaciÃ³n no es perfectamente lineal y existe dispersiÃ³n,
lo que indica que el alcohol es un factor importante pero no el Ãºnico determinante de la calidad.
Vinos con alto contenido de alcohol muestran una mayor concentraciÃ³n en las calidades superiores.

```python
# GrÃ¡fico de violÃ­n para 'fixed acidity' vs 'quality'
plt.figure(figsize=(8, 6))
sns.violinplot(x='quality', y='fixed_acidity', data=df, palette='pastel')
plt.title('Acidez Fija vs Calidad del Vino')
plt.xlabel('Calidad del Vino')
plt.ylabel('Acidez Fija')
plt.show()
```

El grÃ¡fico de violÃ­n entre la acidez fija y la calidad del vino muestra cÃ³mo la distribuciÃ³n de la acidez fija varÃ­a para cada nivel de calidad.
Las formas de los violines sugieren que:
-Para calidades bajas (ej. 3 y 4), la distribuciÃ³n de la acidez fija es mÃ¡s concentrada en valores mÃ¡s bajos.
-A medida que aumenta la calidad (ej. 5, 6, 7, 8), la distribuciÃ³n de la acidez fija tiende a desplazarse hacia valores ligeramente mÃ¡s altos o a tener una mayor dispersiÃ³n.



#DiagnÃ³stico y SelecciÃ³n de Problema:

**CaracterÃ­sticas Principales:**
El dataset incluye variables fisicoquÃ­micas del vino (alcohol, acidez, pH, etc.) y una variable objetivo de calidad (escala 3-8), con mayor concentraciÃ³n en valores intermedios (5-7). El alcohol muestra la correlaciÃ³n positiva mÃ¡s fuerte con la calidad, mientras la acidez volÃ¡til tiene correlaciÃ³n negativa. Se detectan outliers en 'sulphates', pero son valores vÃ¡lidos.

**DesafÃ­os Clave:**
El desequilibrio en las clases de calidad (pocos ejemplos en puntuaciones extremas) complica la predicciÃ³n precisa, especialmente para clasificaciÃ³n. La naturaleza discreta y subjetiva de la calidad plantea dudas sobre abordarla como clasificaciÃ³n o regresiÃ³n. AdemÃ¡s, las mÃºltiples caracterÃ­sticas (11) requieren selecciÃ³n para evitar sobreajuste, y las correlaciones no implican causalidad directa.

**ElecciÃ³n de ClasificaciÃ³n:**
Se prioriza como problema de clasificaciÃ³n (predecir categorÃ­as de calidad) por su aplicabilidad prÃ¡ctica y para identificar factores clave diferenciadores. El desequilibrio se manejarÃ¡ con tÃ©cnicas como oversampling o mÃ©tricas especÃ­ficas (F1-score), aprovechando que el alcohol y acidez volÃ¡til son predictores claros, pero integrando otras variables para capturar la complejidad multifactorial de la calidad del vino.
